 #### There are two approaches below :
 1. To continue using the Categories in "Text" format as they are (for each variable the set of catogories is different but the SCALE is the same)
 2. To conver "Nominal/ORdinal" into -> numerical labels to bring them on a same scale

```python
## Approach number 1 -> continuing with the text categories
happyness1 = happyness[(happyness.Gover!=' ') & (happyness.Hness!=' ') & (happyness.Fear!=' ') ]
```

```python
happyness1[happyness1['Fear'] =='']
print(happyness1.Fear.unique())
```

```python
ax = happyness1['Fear'].value_counts().plot(kind='barh',
                                    figsize=(14,8),
                                    title="Number for each Owner Name")
plt.show()
```

```python
plt.figure(figsize=(10,5))
happyness1['Hness'].value_counts().plot(kind="barh")
plt.show()
happyness1['Gover'].value_counts().plot(kind="barh", color="C2")
plt.show()
happyness1['Fear'].value_counts().plot(kind="barh", color="C3")

#plt.legend(['Happiness','Governance','Fear'])
plt.show()
```

```python
# Approach no.2 -> Converting and bringing all the categories to the same "SCALE"

from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
for i in happyness.columns[2:]:
  happyness[i+'_new'] = ord_enc.fit_transform(happyness[[i]])
happyness.head()
  
```

```python
happyness2 = happyness[(happyness.Gover_new!=0.0) & (happyness.Hness_new!=0.0) & (happyness.Fear_new!=0.0) ]
```

```python
happyness2[happyness2['Fear_new'] ==0.0]
```

```python
happyness2.drop(columns=['Hness','Gover','Fear'],inplace=True,axis=1)
```

```python
# reset_index() gives a column for counting, after groupby uses month and category
# The same logic can be applied to other categorical columns
ctdf = (happyness2.reset_index()
          .groupby(['Date','Hness_new'], as_index=False)
          .count().rename(columns={'index':'ct'}))

fig, ax = plt.subplots()
plt.rcParams["figure.figsize"] = (15,5)
# key gives the group name (i.e. category), data gives the actual values
for key, data in ctdf.groupby('Hness_new'):
    data.plot(x='Date', y='ct', ax=ax, label=key)
```
