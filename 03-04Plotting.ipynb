```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
```

```python
bb=data.DataReader(name='GOOG',data_source='yahoo',start='2002',end='2012')
bb.head()
```

```python
bb.plot(y='High')
```

```python
plt.style.available
plt.style.use('seaborn')
bb.plot(y='High')
```

```python
plt.style.available
```

```python
def sell(volum):
    if volum > 250:
        return 'Good'
    elif 150<volum<250:
        return 'ok'
    else :
        return 'bad'
bb['High'].apply(sell).value_counts().plot(kind='pie',legend=True)
```
