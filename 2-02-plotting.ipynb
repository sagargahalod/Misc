```python
#use OOPS to plot than simple plot as above
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])

ax.plot(x=[2,3,4,5],y=[5,6,7,8])
plt.show()
```

```python
#define data
x=[4,7,8,10]
y=[15,17,18,10]

#ploting subplot on canvas
fig,ax=plt.subplots(figsize=(10,10))

#plot with data
ax.plot(x,y)

#provide label
ax.set(title='Full layout',
      xlabel='x-axis',
      ylabel='y-axis')

#save
fig.savefig('image/sample1.png')

```

```python

df=pd.read_csv("heart-disease.csv")
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df['age'].plot(kind='line');
```

```python
df1 = df[df['age']>50]
df1.plot(kind='scatter',x='age',
       y='chol',c='target')
```

```python
#Object oriented
fig,ax=plt.subplots(figsize=(10,5))
df1.plot(kind='scatter',x='age',
       y='chol',c='target',
        ax=ax)
```

```python
#complete OOP
fig,ax=plt.subplots(figsize=(10,6))
scatter=plt.scatter(x=df['age'],y=df['chol'],
                   c=df['target'])
ax.set(title='analysis',xlabel='age',ylabel='chole')
ax.axhline(df1['chol'].mean(),linestyle='--')
```
